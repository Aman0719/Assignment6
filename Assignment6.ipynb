```python
#Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

Ans- 

#Class
In object-oriented programming, a class is a blueprint for creating objects that defines a set of attributes and behaviors. 
I simple words, Classes are user-defined data types that act as the blueprint for individual objects, attributes and methods

#Object
An object is an instance of a class that has its own unique state and behavior, and can interact with other objects or the environment.
Simply, Objects are instances of a class created with specifically defined data

A class typically consists of data members (attributes) and member functions (methods) that operate on the data members. 
The attributes define the state of the object, while the methods define the behavior of the object.
```

```python
#example of class

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.speed = 0

    def accelerate(self):
        self.speed += 10

    def brake(self):
        self.speed -= 10

    def get_speed(self):
        return self.speed

#In this example,the Car class has four attributes (make, model, year, and speed) and three methods (accelerate, brake, and get_speed).
#The __init__ method is a special method that is called when a new object of the Car class is created. 
#It initializes the attributes of the object with the values provided as arguments.
```

```python
#example of object

To create an object of the Car class,

my_car = Car("Toyota", "Innova", 2022)

#In this example, we create a new object my_car of the Car class with the make "Toyota", model "Innova", and year 2022. 
#We can then use the methods of the my_car object to accelerate, brake, and get the current speed:

my_car.accelerate()  
my_car.accelerate()  
my_car.brake()       
print(my_car.get_speed()) 

```

```python
# Q2. Name the four pillars of OOPs.

Ans-

The four pillars of object-oriented programming (OOP) are:

1. Encapsulation: 
This refers to the practice of grouping related data and behavior (methods) into a single unit called a class. 
Encapsulation hides the implementation details of the class from other parts of the program, and provides a clean interface for interacting with the class. 
This helps to achieve data abstraction, information hiding, and modularity.

2. Inheritance: 
This is the mechanism by which one class (the subclass or derived class) inherits the properties (attributes and methods) of another class (the superclass or base class). 
Inheritance allows for code reuse, reduces code duplication, and facilitates the creation of a hierarchy of related classes.

3. Polymorphism:
This refers to the ability of an object to take on multiple forms or behaviors depending on the context in which it is used. 
Polymorphism allows for more flexible and extensible code by enabling the same interface to be used with different types of objects. 
Polymorphism can be achieved through method overloading, method overriding, or operator overloading.

4. Abstraction: 
This refers to the practice of representing complex real-world entities as simplified models in software. 
Abstraction focuses on the essential features of an entity and ignores the non-essential details. 
Abstraction enables programmers to create reusable software components that can be used in different contexts without modification. 
Abstraction can be achieved through abstract classes and interfaces.
```

```python
# Q3. Explain why the __init__() function is used. Give a suitable example.

Ans-
In object-oriented programming, the __init__() function is a special method that is called when an object of a class is created. 
It is used to initialize the attributes of the object with the values provided as arguments. 
The __init__() method is optional, but it is commonly used to ensure that objects are properly initialized when they are created.

Example in next cell
```

```python
class Person:
    def __init__(self, name, age, occupation):
        self.name = name
        self.age = age
        self.occupation = occupation
    
    def introduce(self):
        print(f"My name is {self.name}. I am {self.age} years old, and I work as a {self.occupation}.")

        
#In this example, we define a Person class with an __init__() method that initializes the name, age, and occupation attributes of the object with the values provided as arguments. 
#We also define a introduce() method that prints a message introducing the person.

person1 = Person("Aman", 27, "Data_Scientist")

person1.introduce()
```

```python
Q5. What is inheritance? Give an example for each type of inheritance.

Ans-

Inheritance is a mechanism in object-oriented programming where a new class is created from an existing class, inheriting all the properties and methods of the parent class. 
The new class is called the subclass or derived class, and the existing class is called the superclass or base class. 
The purpose of inheritance is to promote code reuse and make it easier to create new classes with similar functionality to existing classes.

There are four types of inheritance:

1. Single Inheritance: 
In single inheritance, a subclass inherits from a single superclass. 
For example, a class "Car" can be a superclass, and a class "Sedan" can be a subclass that inherits from "Car". 
The Sedan class would inherit all the properties and methods of the Car class.

2. Multilevel Inheritance: 
In multilevel inheritance, a subclass inherits from a superclass, which itself is a subclass of another superclass. 
For example, a class "Vehicle" can be a superclass, a class "Car" can be a subclass that inherits from "Vehicle", and a class "Sedan" can be a subclass that inherits from "Car". 
The Sedan class would inherit all the properties and methods of both the Car and Vehicle classes.

3. Hierarchical Inheritance: 
In hierarchical inheritance, multiple subclasses inherit from a single superclass. 
For example, a class "Animal" can be a superclass, and classes "Cat" and "Dog" can be subclasses that both inherit from "Animal". 
Both the Cat and Dog classes would inherit all the properties and methods of the Animal class.

4. Multiple Inheritance: 
In multiple inheritance, a subclass inherits from multiple superclasses. 
For example, a class "Athlete" can be a superclass, and classes "Runner" and "Swimmer" can be subclasses that both inherit from "Athlete". 
The subclass "Triathlete" could then inherit from both the "Runner" and "Swimmer" classes, effectively inheriting properties and methods from multiple classes.
```

```python
#Example of Single Inheritance

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started.")

    def stop_engine(self):
        print("Engine stopped.")

class Sedan(Car):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def open_trunk(self):
        print("Trunk opened.")
```

```python
#Example of Multilevel Inheritance:

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started.")

    def stop_engine(self):
        print("Engine stopped.")

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def honk_horn(self):
        print("Honk honk!")

class Sedan(Car):
    def __init__(self, make, model, year, num_doors, trunk_capacity):
        super().__init__(make, model, year, num_doors)
        self.trunk_capacity = trunk_capacity

    def open_trunk(self):
        print("Trunk opened.")

    def close_trunk(self):
        print("Trunk closed.")
```

```python
#Example of Hierarchical Inheritance:

class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def make_sound(self):
        pass

class Cat(Animal):
    def make_sound(self):
        print("Meow")

class Dog(Animal):
    def make_sound(self):
        print("Woof")
```

```python
#Example of Multiple Inheritance:

class Athlete:
    def __init__(self, name, age, sport):
        self.name = name
        self.age = age
        self.sport = sport

    def warm_up(self):
        print("Warming up...")

class Runner(Athlete):
    def run(self):
        print("Running...")

class Swimmer(Athlete):
    def swim(self):
        print("Swimming...")
```
